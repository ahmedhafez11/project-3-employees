```python
with open('employees.csv', 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['id', 'name', 'email', 'gender', 'active'])
    writer.writeheader() # Inserting column names at the start

    for e in employees['data']:
        id_ = e['id']
        name = e['name']
        email = e['email']
        gender = e['gender']
        active = True if e['status'] == 'active' else False
        
        writer.writerow(
            {
                'id' : id_,
                'name' : name,
                'email' : email,
                'gender' : gender,
                'active' : active

            }
        )
```
